#### A few key concepts to understand before starting this tutorial

```python
# Import modules which supports typed annotations
from typing import TypedDict

# Define typed dictionary
class Movie(TypedDict):
    """
    Represents a movie record with a fixed schema.
    Notes:
        - TypedDict is used for type hinting only.
        - It does NOT enforce types at runtime.
        - Useful for static type checking and IDE support.
    """
    name: str
    year: int


movie_01 = Movie(name="3 Idiots", year=2006)
movie_02 = Movie(name="3 Idiots", year="some_string") # This should fail but output is fine

print(movie_01)
print(movie_02)
```

```python
# Using Union for Multiple Accepted Input Types
from typing import Union

def square(x: Union[int, float]) -> float:
    """
    Notes:
        - Union[int, float] indicates that 'x' can be either an int or a float.
        - This is a type hint only; Python does not enforce it at runtime.
        - Invalid types may still cause runtime errors depending on operations.
    """
    return x * x

print(square(5))
print(square(5.2))

# Lets try and print square of a string and catch the error
try:
    print("Square of string:", square("some_string"))
except TypeError as e:
    print("Runtime error:", e)
```

```python
# Using optional
from typing import Optional

def nice_massage(name: Optional[str]) -> None:
    if name is None:
        print("Hey, do I know you?")
    else:
        print(f"Hi there, {name}")

nice_massage(None)
nice_massage("Vikas")
```

```python
# Understanding Lambda (A small inline function for quick operations)
#
square = lambda x: x * x # lambda arguments: expression 
print(f"square of 5 is {square(5)}")

volume = lambda l, b, h: l * b * h # multi arguments
print(f"The calculated volume is {volume(3, 4, 5)}")

nums = [1, 2, 3, 4]
squares = list(map(lambda x: x * x, nums )) # map(function, iterable)
print(f"Square of {nums}  is {squares}")
print(f"output of map is {map(lambda x: x * x, nums )}") #Output not useful
```
